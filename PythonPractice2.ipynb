# Working with an existing database i.e reading, indexing, accessing, selecting from an existing database


When you load a dataset using Pandas, it automatically creates an index column if one doesn't exist. This index starts from 0 and increments by 1 for each row. For example, when you use the head() function to display the first 5 rows of your dataset, you'll notice a column to the left of your data labeled with numbers like 0, 1, 2, etc. This is the index created by Pandas, which helps in uniquely identifying each row.

However, if your CSV file already contains an index column and you want to use that existing index instead of letting Pandas create a new one, you can specify this when reading the CSV file. This prevents Pandas from generating an additional index column, keeping your data as it originally is in the file.

To achieve this, you can use the index_col parameter in the pd.read_csv() function. Here's how you can do it:

pd.read_csv("/content/sample_data/GlobalData.csv", index_col=0)

In this example, the first column in the CSV file (index_col= 0) will be used as the index for the DataFrame, and no extra index column will be created by Pandas.

```python
df=pd.read_csv("/content/sample_data/GlobalData.csv") #reading dataset and assining to a variable. df is our new dataframe.
df.shape #this will give how much rows n columns are in our dataset.Shape is an attribute
#reading the data from csv using head()method, which will display 1st five rows from the dataset
df.head()
```
